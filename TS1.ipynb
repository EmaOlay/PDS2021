```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
def mi_funcion_sen( vmax,dc,ff,ph,nn,fs ):
    ##Grafico una senoidal
    ts = 1/fs # tiempo de muestreo
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (nn-1)*ts, nn)#.flatten()
    
    aux=vmax * np.sin( 2*np.pi*ff*tt + ph ) + dc
    # para concatenar horizontalmente es necesario cuidar que tengan iguales FILAS
    ##x = np.hstack([x, aux.reshape(N,1)] )
    plt.figure(1)
    line_hdls = plt.plot(tt, aux)
    plt.title('Señal: Senoidal')
    plt.xlabel('tiempo [segundos]')
    plt.ylabel('Amplitud [V]')
    
    # presentar una leyenda para cada tipo de señal
    #axes_hdl = plt.gca()
    
    # este tipo de sintaxis es *MUY* de Python
    notita = mpatches.Patch(label=str(ff)+'hz')
    plt.legend(handles=[notita])
    
    plt.show()
```

```python
mi_funcion_sen( vmax = 3, dc = 2, ff = 2, ph=np.pi, nn = 500, fs = 500)
```

```python
def mi_funcion_cuadrada( vmax=1,dc=0,duty=0.5,nn=1000,fs=1000 ):
    ##la funcion pone en Vmax +dc a la señal durante el duty cicle
    
    ts = 1/fs # tiempo de muestreo
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (nn-1)*ts, nn)
    aux=np.copy(tt)
    aux.fill(0)
    i=len(tt)
    print(i)
    
    for x in range(i):
      if x<=(i*duty):
        aux[x]= vmax+dc
      else:
        aux[x]=dc
    
    ### para concatenar horizontalmente es necesario cuidar que tengan iguales FILAS
    
    plt.figure(1)
    line_hdls = plt.plot(tt, aux)
    plt.title('Señal: Cuadrada montada sobre dc')
    plt.xlabel('tiempo [segundos]')
    plt.ylabel('Amplitud [V]')
    
    notita = mpatches.Patch(label='duty'+str(duty)+'%')
    plt.legend(handles=[notita])
    
    plt.show()
```

```python
mi_funcion_cuadrada( vmax=1,dc=0,duty=0.6,nn=1000,fs=1000 )
```
