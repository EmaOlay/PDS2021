```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
N = 20 # muestras
B = 1 # bits
##limite sup 2**(B-1)
##limite inf 2**(B-1)+1
## pruebo el redondeo std => round

##vector tiempo de 0 a 1, no incluido, con N saltos
t = np.linspace(0, 1, N, endpoint=False)


#senoidal de amplitud 1 y 1 hz de frecuencia
amp=2**(B-1) #amplitud en volts
freq=1 #frecuencia en Hz
fase=0 #radianes
seno = amp*np.sin(2 * np.pi * freq * t + fase)
muestreada=np.round(seno)

plt.plot(t,seno)
plt.stem(t, muestreada)
plt.show()
```

```python
Error1=muestreada-seno

plt.stem(t, Error1)
plt.show()
```

# Conclusiones
El error no supera el valor 0.5 por motivos propios de la funcion round probablemente no tenga sentido hacer el grafico...<br>
Pero nos permite obtener una primer mirada entre los errores a ver si hay correlacion entre ellos...

```python
corr = sig.correlate(Error1, Error2, mode='same')

plt.plot(t, corr)
plt.show()
```
